# Load full dataset

```python
import pandas as pd

full_dataset = pd.read_csv('Absenteeism_at_work.csv', sep=';')
full_dataset.head()
```

```python
# Drop columns
full_dataset = full_dataset.drop(['ID', 'Reason for absence', 'Month of absence', 'Day of the week', 'Seasons', 'Service time'], axis=1)
full_dataset.head()
```

# Separate training and predict csv s

```python
# Generate a train_dataset.csv removing the last 70 rows
train_dataset = full_dataset[:-70]
train_dataset.to_csv('train_dataset.csv', index=False)
```

```python
# Generate a test_dataset.csv containing only the last 70 rows
# test_dataset = full_dataset[-100:-30]
# test_dataset.to_csv('test_dataset.csv', index=False)
```

```python
# Generate a predict_dataset.csv containing only the last 70 rows and dropping the 'Absenteeism time in hours' column
predict_dataset = full_dataset[-70:]
predict_dataset = predict_dataset.drop(['Absenteeism time in hours'], axis=1)
predict_dataset.to_csv('predict_dataset.csv', index=False)
```
