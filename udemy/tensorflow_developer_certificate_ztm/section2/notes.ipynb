# What is deep learning?
It's a type of machine learning based on artificial neural networks in which multiple layers of processing are used to extract progressively higher level features from data.

Is to get things (data) turn it into numbers and find patterns in those numbers.

Deep learning can be considered a subfield of machine learning. And machine learning is a subfield of artificial intelligence.

## Traditional programming

You collect inputs, code some rules and then get the output.

## Machine learning algorithm

You start with inputs and outputs and them discover the rules.

# Why use machine learning?
For a complex problem we can't think of all the rules to program it, but a ML system can figure out those rules.

ML may be a good solution when:
* The problem has too many rules.
* The environment is continually changing.
* When we have to discover insights within large collections of data.

Scenarios were ML may not be a good approach:
* When you need explainability.
* When the traditional approach is a better option.
* When errors are unacceptable.
* When you don't have much data.

When to use machine learning: **On structured data.**

When to use deep learning: **On unstructured data.**

# What are neural networks?
A network of neurons that consist of layers of interconnected nodes that process and transmit information.

We get inputs (text, audio, images, etc), turn them into numbers and then we feed them into the neural network, that learn patterns in those numbers. The network outputs numbers and we have to interpret those numbers.

## Types of learning
* Supervised learning: data and labels
* Semi-supervised learning: data and some labels
* Unsupervised learning: you have data but no labels
* Transfer learning: use a model trained on a problem that is similar to your actual problem

# Applications where deep learning is already being used for:
* **Recommendation (videos, movies, musics, etc):** classification/regression
* **Translation:** seq2seq
* **Speech recognition:** seq2seq
* **Computer vision:** classification/regression
* **Natural language processing:** classification/regression

# TensorFlow
* Is an end to end platform for developing and deploying machine learning/deep learning models: preprocess data, train, evaluate, deploy.
* TensorFlow makes it possible to run the code on different types of hardware (CPU, GPU, TPU).
* You can access many pre-build models (transfer learning).
