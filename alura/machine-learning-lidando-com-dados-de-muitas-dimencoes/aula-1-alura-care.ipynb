```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from numpy import random

SEED = 123143
random.seed(SEED)

valores_exames = resultados_exames.drop(columns=['id', 'diagnostico'])
diagnostico = resultados_exames.diagnostico
# remover a coluna 33 (tem muitos valores nulos)
valores_exames_v1 = valores_exames.drop(columns=['exame_33'])

treino_x, teste_x, treino_y, teste_y = train_test_split(valores_exames_v1, diagnostico, test_size=0.3)


# instanciar objeto do modelo
classificador = RandomForestClassifier(n_estimators=100)
# ajustar modelo aos dados
classificador.fit(treino_x, treino_y)
print('Resultado da classificaÃ§Ã£o: ', (classificador.score(teste_x, teste_y) * 100))
```

```python
# compara resultado obtido com o dummyClassifier
from sklearn.dummy import DummyClassifier
SEED = 123143
random.seed(SEED)
classificador_bobo = DummyClassifier(strategy='most_frequent')
classificador_bobo.fit(treino_x, treino_y)
print('Resultado da classificaÃ§Ã£o boba: ', (classificador_bobo.score(teste_x, teste_y) * 100))
```

```python
def grafico_violino(valores, inicio, fim):
    dados_plot = pd.concat([diagnostico, valores.iloc[:, inicio:fim]], axis=1)
    dados_plot = pd.melt(dados_plot, id_vars='diagnostico', var_name='exames', value_name='valores')
    plt.figure(figsize=(10, 10))
    sns.violinplot(x='exames', y='valores', hue='diagnostico', data=dados_plot, split=True)
    plt.xticks(rotation=90)

# temos cerca de 30 exames (colunas), antes plotamos os 10 primeiros e agora os 10 seguintes
# e em seguida os Ãºltimos 10 exames
grafico_violino(valores_exames_v2, 10, 20)
```

```python
# Remover as colunas com valores constantes
valores_exames_v3 = valores_exames_v2.drop(columns=['exame_4', 'exame_29'])
valores_exames_v3
```

```python
import random
def classificar(valores):
    SEED = 1234
    random.seed(SEED)
    treino_x, teste_x, treino_y, teste_y = train_test_split(valores, diagnostico, test_size=0.3)
    # instanciar objeto do modelo
    classificador = RandomForestClassifier(n_estimators=100)
    # ajustar modelo aos dados
    classificador.fit(treino_x, treino_y)
    print('Resultado da classificaÃ§Ã£o: ', (classificador.score(teste_x, teste_y) * 100))

classificar(valores_exames_v3)
```

```python
matriz_correlacao = valores_exames_v3.corr()
plt.figure(figsize=(17, 15))
sns.heatmap(matriz_correlacao, annot=True, fmt='.1f')
```

```python
# filtrar as correlaÃ§Ãµes maiores que 0.9
matriz_correlacao_v1 = matriz_correlacao[matriz_correlacao > 0.99]
matriz_correlacao_v1
```

```python
# fazer a soma, para identificar colunas de alta correlaÃ§Ã£o
matriz_correlacao_v2 = matriz_correlacao_v1.sum()
matriz_correlacao_v2
```

```python
# separar as variaveis altamente correlacionadas
variaveis_correlacionadas = matriz_correlacao_v2[matriz_correlacao_v2>1]
variaveis_correlacionadas
```

```python
# excluir as features altamente correlacionadas
valores_exames_v4 = valores_exames_v3.drop(columns=variaveis_correlacionadas.keys())
valores_exames_v4
# aqui, na verdade nÃ£o se deve excluir todas as variaveis correlacionadas, como elas
# se correlacionam em pares, basta exluir uma das variaveis do par.
```

```python
# medir a accuracia com o novo dataset
classificar(valores_exames_v4)
```

```python
# ajustar dataset excluindo apenas uma das variÃ¡veis do par
# exame 1 se correlaciona com exame 3, entÃ£o removemos a coluna exame 3
# exame 22 se correlaciona com exame 24, entÃ£o removemos a coluna exame 24
valores_exames_v5 = valores_exames_v3.drop(columns=['exame_3', 'exame_24'])
# classificar novamente pra observar as mudanÃ§as
classificar(valores_exames_v5)
```
