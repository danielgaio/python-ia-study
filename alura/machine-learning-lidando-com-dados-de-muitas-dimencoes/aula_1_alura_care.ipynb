### Selecionar as N melhores features automaticamente

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
# chi2 é uma função que calcula o valor de chi2 para cada feature.
# só aceita valores positivos

k_best = SelectKBest(chi2, k=5)

# vamos usar valores_exames_v1 antes da normalização pra não ter valores negativos
valores_exames_v6 = valores_exames_v1.drop(columns=['exame_3', 'exame_4', 'exame_24', 'exame_29'])

SEED = 1234
random.seed(SEED)

treino_x, teste_x, treino_y, teste_y = train_test_split(valores_exames_v6, diagnostico, test_size=0.3)

k_best.fit(treino_x, treino_y)

treino_k_best = k_best.transform(treino_x)
teste_k_best = k_best.transform(teste_x)
```

```python
treino_k_best.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
classificador = RandomForestClassifier(n_estimators=100, random_state=1234)
classificador.fit(treino_k_best, treino_y)
print('Resultado da classificação: %.2f%%' %(classificador.score(teste_k_best, teste_y)*100))
# aqui utilizamos apenas 5 features e o resultados ainda é satisfatório, ou seja,
# em termos de exames, onde antes era nescessário realizar mais de 30 exames
# agora é preciso apenas 5 para ter uma predição equivalente do tipo de câncer
```

### Avaliando o modelo pela matriz de confusão

```python
from sklearn.metrics import confusion_matrix

conf_matrix = confusion_matrix(teste_y, classificador.predict(teste_k_best))
conf_matrix
```

```python
# plotar a matriz de confusão com o seaborn heatmap
plt.figure(figsize=(9, 7))
sns.set(font_scale=2)
sns.heatmap(conf_matrix, annot=True, fmt='d').set(xlabel='Predito', ylabel='Real')
# nesse caso 1 é cancer maligno, então a intersecção 1/0 da matriz apresenta as predições onde
# o paciente tinha cancer mas o modelo classificou como não tendo. Neste problema este quadrante
# é especialmente importante pois um diagnostico incorreto pode fazer com que o paciente perca
# tempo de tratamento o que pode ser fatal, então deve-se chegar o mais próximo de zero para
# esse quadrante de falsos negativos.
```

### Recursive Feature Elimination - RFE

```python
from sklearn.feature_selection import RFE

SEED = 1234
random.seed(SEED)

treino_x, teste_x, treino_y, teste_y = train_test_split(valores_exames_v6, diagnostico, test_size=0.3)

classificador = RandomForestClassifier(n_estimators=100, random_state=1234)
classificador.fit(treino_x, treino_y)

rfe_selector = RFE(estimator=classificador, n_features_to_select=2, step=1)
rfe_selector.fit(treino_x, treino_y)
treino_rfe = rfe_selector.transform(treino_x)
teste_rfe = rfe_selector.transform(teste_x)

classificador.fit(treino_rfe, treino_y)
print('Resultado da classificação: %.2f%%' %(classificador.score(teste_rfe, teste_y)*100))
```

```python
conf_matrix = confusion_matrix(teste_y, classificador.predict(teste_rfe))

plt.figure(figsize=(9, 7))
sns.set(font_scale=2)
sns.heatmap(conf_matrix, annot=True, fmt='d').set(xlabel='Predito', ylabel='Real')
```

### RFECV - Recursive Feature Elimination Cross Validation

```python
from sklearn.feature_selection import RFECV

SEED = 1234
random.seed(SEED)

treino_x, teste_x, treino_y, teste_y = train_test_split(valores_exames_v6, diagnostico, test_size=0.3)

classificador = RandomForestClassifier(n_estimators=100, random_state=1234)
classificador.fit(treino_x, treino_y)

rfecv_selector = RFECV(estimator=classificador, cv=5, step=1, scoring='accuracy')
rfecv_selector.fit(treino_x, treino_y)

treino_rfecv = rfecv_selector.transform(treino_x)
teste_rfecv = rfecv_selector.transform(teste_x)

classificador.fit(treino_rfecv, treino_y)
print('Resultado da classificação: %.2f%%' %(classificador.score(teste_rfecv, teste_y)*100))

conf_matrix = confusion_matrix(teste_y, classificador.predict(teste_rfecv))
plt.figure(figsize=(9, 7))
sns.set(font_scale=2)
sns.heatmap(conf_matrix, annot=True, fmt='d').set(xlabel='Predito', ylabel='Real')
```

```python
# exibir quantas features foram usadas
rfecv_selector.n_features_
```

```python
# visualizar quais são as features
print(rfecv_selector.support_)
print(len(rfecv_selector.support_))
print(rfecv_selector.n_features_)
# exibir array com score de cada feature
# print(rfecv_selector.grid_scores_)  # deprecated
print(rfecv_selector.cv_results_)
# ver quais são essas colunas em treino x
treino_x.columns[rfecv_selector.support_]
```

```python
rfecv_selector.cv_results_['mean_test_score']
```

```python
# plotar gráfico para entender porque essas features foram selecionadas
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 7))
plt.xlabel('Número de exames')
plt.ylabel('Acurácia')
plt.plot(range(1, len(rfecv_selector.cv_results_['mean_test_score']) + 1), rfecv_selector.cv_results_['mean_test_score'])
plt.show()
# o gráfico mostra a relação entre a quantidade de exames sendo utilizada e a acurácia obtida para cada quantidade
```
